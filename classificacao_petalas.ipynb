```python
from sklearn import datasets

iris = datasets.load_iris()

print(iris.feature_names)
print(iris.target_names)
print(iris.data[0])
print(iris.target[0])

for i in range(len(iris.target)):
    print
```

```python
x = iris.data
y = iris.target

y
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(x)
x = scaler.transform(x)
```

```python
from sklearn.preprocessing import OneHotEncoder

# Substitua 'categorical_features' por 'handle_unknown' e 'categories'
# handle_unknown='ignore' instrui o codificador a ignorar valores desconhecidos durante a etapa de transformação
# categories='auto' irá inferir as categorias automaticamente dos dados de treinamento
# Em vez de 'categorical=[0]', use 'categories' para especificar as categorias
# e 'sparse_output=False' para obter um array denso diretamente
onehot = OneHotEncoder(categories='auto', sparse_output=False, handle_unknown='ignore')
y = y.reshape(-1, 1)
y = onehot.fit_transform(y)
y
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np

neuronios_entrada = x.shape[1]
neuronios_saida = y.shape[1]
neuronios_oculta = int(np.ceil((y.shape[1] + x.shape[1]) / 2))

model = keras.Sequential()
model.add(layers.Dense(neuronios_oculta, input_dim=neuronios_entrada, activation='relu'))
model.add(layers.Dense(neuronios_saida, activation='softmax'))
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Treina o modelo e guarda o histórico
history = model.fit(x_train, y_train, epochs=100, batch_size=32, validation_data=(x_test, y_test))

# Avalia o modelo nos dados de teste
loss, accuracy = model.evaluate(x_test, y_test, verbose=0)
print(f"Loss no conjunto de teste: {loss:.4f}")
print(f"Acurácia no conjunto de teste: {accuracy:.4f}")

# Imprime gráficos do histórico de treinamento
import matplotlib.pyplot as plt

plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Acurácia do Modelo')
plt.ylabel('Acurácia')
plt.xlabel('Época')
plt.legend(['Treinamento', 'Validação'], loc='upper left')
plt.show()
```
